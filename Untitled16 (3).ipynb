```python
# 1️⃣ Montar Google Drive (si no está montado)
from google.colab import drive
import os

drive.mount('/content/drive')

# 📂 Ruta corregida del dataset
dataset_path = "/content/drive/MyDrive/IAn Abaj Dataseet"

# 🖼️ Preprocesamiento de imágenes
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# Tamaño de imágenes y batch size
image_size = (224, 224)
batch_size = 16

# Generador de imágenes con aumento de datos
datagen = ImageDataGenerator(
    rescale=1.0/255,  # Normalización
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # 80% entrenamiento, 20% validación
)

train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# 4️⃣ Cargar MobileNetV2 preentrenado (sin la capa de salida)
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Congelamos las capas base

# 🔗 Agregar capas personalizadas
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
output_layer = Dense(train_generator.num_classes, activation='softmax')(x)

# 🚀 Crear modelo
model = Model(inputs=base_model.input, outputs=output_layer)

# ⚙️ Compilar modelo
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 🎯 Entrenar modelo
model.fit(train_generator, validation_data=val_generator, epochs=10)

# 💾 Guardar modelo entrenado
model.save("/content/drive/MyDrive/modelo_IA_agricola.h5")
print("✅ Modelo guardado correctamente en Google Drive.")
```

```python
import os
import shutil
import random

# Define la ruta base del dataset
base_dir = "/content/drive/MyDrive/IAn Abaj Dataseet"  # Ajusta esta ruta si es necesario

# Carpetas principales
cultivos = ['FRIJOL', 'CALABAZA', 'CACAO', 'CAMOTE', 'MAIZ', 'YUCA']

# Crear las carpetas 'train', 'test' y 'val' si no existen
for folder in ['train', 'test', 'val']:
    for cultivo in cultivos:
        os.makedirs(os.path.join(base_dir, folder, cultivo), exist_ok=True)

# Mover las imágenes a las carpetas correspondientes
for cultivo in cultivos:
    cultivo_dir = os.path.join(base_dir, cultivo)
    imagenes = os.listdir(cultivo_dir)

    # Barajar las imágenes para distribuirlas aleatoriamente
    random.shuffle(imagenes)

    # Dividir en 70% train, 15% val, 15% test
    train_split = int(0.7 * len(imagenes))
    val_split = int(0.15 * len(imagenes))

    # Mover las imágenes
    for i, imagen in enumerate(imagenes):
        if i < train_split:
            destino = os.path.join(base_dir, 'train', cultivo, imagen)
        elif i < train_split + val_split:
            destino = os.path.join(base_dir, 'val', cultivo, imagen)
        else:
            destino = os.path.join(base_dir, 'test', cultivo, imagen)

        # Mover la imagen
        shutil.move(os.path.join(cultivo_dir, imagen), destino)

print("¡Dataset organizado correctamente!")
```

```python
# Directorio donde están las carpetas train, val y test
base_dir = "/content/drive/MyDrive/IAn Abaj Dataseet"

# Generadores de imágenes
train_datagen = ImageDataGenerator(
    rescale=1./255,          # Normalización de los valores de píxeles
    shear_range=0.2,         # Aleatoriza los ángulos de las imágenes
    zoom_range=0.2,          # Aumento de zoom
    horizontal_flip=True     # Aumentar mediante voltear las imágenes horizontalmente
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Cargar los datos desde las carpetas
train_generator = train_datagen.flow_from_directory(
    os.path.join(base_dir, 'train'),
    target_size=(150, 150),   # Cambiar tamaño de las imágenes
    batch_size=32,
    class_mode='categorical'  # Categoría (como tenemos varias clases)
)

val_generator = val_datagen.flow_from_directory(
    os.path.join(base_dir, 'val'),
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)
```

```python
# Construir el modelo CNN
model = models.Sequential([
    # Capa convolucional y de agrupamiento
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Aplanar y añadir capas densas
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),  # Evitar sobreajuste con Dropout
    layers.Dense(len(train_generator.class_indices), activation='softmax')  # Una salida por clase
])

# Compilar el modelo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Resumen del modelo
model.summary()
```

```python
import os
import cv2
import numpy as np

# Directorio donde están las imágenes (ajusta la ruta si es necesario)
directorio_imagenes = "/content/drive/MyDrive/IAn Abaj Dataseet/train"

# Listas para almacenar las imágenes y las etiquetas
X_images = []
y_labels = []

# Extensiones de imágenes válidas
extensiones_validas = ['.jpg', '.jpeg', '.png']

# Función para procesar las carpetas y subcarpetas recursivamente
def cargar_imagenes(directorio):
    for folder_name in os.listdir(directorio):
        folder_path = os.path.join(directorio, folder_name)

        if os.path.isdir(folder_path):  # Si es una carpeta
            print(f"\nProcesando imágenes en la carpeta: {folder_name}")

            # Recorrer archivos dentro de esta carpeta o subcarpeta
            for root, _, files in os.walk(folder_path):
                for img_name in files:
                    img_path = os.path.join(root, img_name)

                    # Imprimir el nombre completo del archivo para depuración
                    print(f"Verificando archivo: {img_path}")

                    # Comprobar si el archivo tiene una extensión válida
                    if not any(img_name.lower().endswith(ext) for ext in extensiones_validas):
                        print(f"Archivo no válido (no es imagen): {img_name}")
                        continue  # Saltar archivos no válidos

                    # Verificar si el archivo realmente es una imagen
                    img = cv2.imread(img_path)
                    if img is None:
                        print(f"Error al cargar la imagen: {img_path}. Esto podría indicar que la imagen está corrupta o no es válida.")
                        continue  # Saltar imágenes no válidas

                    # Redimensionar la imagen
                    img = cv2.resize(img, (224, 224))  # Redimensionar la imagen a 224x224
                    X_images.append(img)
                    y_labels.append(folder_name)  # Usar el nombre de la carpeta como la etiqueta

# Llamamos a la función de carga de imágenes
cargar_imagenes(directorio_imagenes)

# Imprimir cuántas imágenes se cargaron
print(f"\nTotal de imágenes cargadas: {len(X_images)}")

# Verificar algunas imágenes cargadas para depurar
if len(X_images) > 0:
    print(f"Primera imagen cargada: {X_images[0].shape}")  # Mostrar el tamaño de la primera imagen cargada
else:
    print("No se cargaron imágenes.")
```

```python
from PIL import Image
import os

def check_and_convert_images(directory):
    # Recorre todas las carpetas
    for folder in os.listdir(directory):
        folder_path = os.path.join(directory, folder)
        if os.path.isdir(folder_path):
            for filename in os.listdir(folder_path):
                file_path = os.path.join(folder_path, filename)
                # Asegúrate de que el archivo sea una imagen
                if filename.lower().endswith(('png', 'jpg', 'jpeg')):
                    try:
                        with Image.open(file_path) as img:
                            img.verify()  # Verifica si la imagen está corrupta
                            if img.format not in ['JPEG', 'PNG']:  # Si no es JPEG ni PNG
                                new_file_path = file_path.rsplit('.', 1)[0] + '.png'
                                img = Image.open(file_path)
                                img.save(new_file_path, 'PNG')
                                os.remove(file_path)  # Eliminar el archivo original
                                print(f'Convertido {filename} a PNG')
                    except Exception as e:
                        print(f'Error al verificar la imagen {filename}: {e}')

# Verificar y convertir imágenes en la carpeta de entrenamiento y validación
check_and_convert_images('/content/drive/MyDrive/IAn Abaj Dataseet/train')
check_and_convert_images('/content/drive/MyDrive/IAn Abaj Dataseet/val')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Rutas a las carpetas
train_dir = '/content/drive/MyDrive/IAn Abaj Dataseet/train'
val_dir = '/content/drive/MyDrive/IAn Abaj Dataseet/val'
test_dir = '/content/drive/MyDrive/IAn Abaj Dataseet/test'

# Definir tamaño de las imágenes
IMG_SIZE = (224, 224)

# Cargar los datasets
train_ds = image_dataset_from_directory(
    train_dir,
    image_size=IMG_SIZE,
    batch_size=32,  # Tamaño del lote
    label_mode='categorical'  # Etiquetas en formato categórico
)

val_ds = image_dataset_from_directory(
    val_dir,
    image_size=IMG_SIZE,
    batch_size=32,
    label_mode='categorical'
)

test_ds = image_dataset_from_directory(
    test_dir,
    image_size=IMG_SIZE,
    batch_size=32,
    label_mode='categorical'
)

# Inspeccionar las clases
print("Clases en entrenamiento:", train_ds.class_names)
print("Clases en validación:", val_ds.class_names)
print("Clases en prueba:", test_ds.class_names)
```

```python
from tensorflow.keras import layers, models

# Crear el modelo
model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(224, 224, 3)),  # Normalización de imágenes
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(6, activation='softmax')  # 6 clases en tu dataset
])

# Compilar el modelo
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Resumen del modelo
model.summary()
```

```python
import gradio as gr
import tensorflow as tf
import numpy as np
from PIL import Image as PILImage

# Cargar el modelo .tflite
model_path = '/content/drive/MyDrive/IAn Abaj/mi_modelo.tflite'
interpreter = tf.lite.Interpreter(model_path=model_path)
interpreter.allocate_tensors()

# Función de predicción
def predict(image_path):
    # Convertir la imagen a un formato que el modelo pueda entender
    img = PILImage.open(image_path)
    img = img.resize((224, 224))  # Asegurarse de que la imagen tenga el tamaño adecuado
    img_array = np.array(img) / 255.0  # Normalizar los valores de píxeles
    img_array = np.expand_dims(img_array, axis=0)  # Agregar una dimensión para el batch size

    # Realizar la predicción
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    interpreter.set_tensor(input_details[0]['index'], img_array.astype(np.float32))
    interpreter.invoke()

    # Obtener la predicción
    output_data = interpreter.get_tensor(output_details[0]['index'])
    return output_data

# Interfaz de usuario con Gradio
interface = gr.Interface(fn=predict,
                         inputs=gr.Image(type="filepath"),  # Cambiado a "filepath"
                         outputs=gr.Label(num_top_classes=3))  # Cambiado a gr.Label

# Lanzar la interfaz
interface.launch(share=True)
```

```python
import gradio as gr
import tensorflow as tf
import numpy as np
from PIL import Image as PILImage

# Cargar el modelo TFLite
interpreter = tf.lite.Interpreter(model_path="/content/drive/MyDrive/IAn Abaj/mi_modelo.tflite")
interpreter.allocate_tensors()

# Obtener detalles de las entradas y salidas del modelo
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Función de predicción
def predict(image):
    # La imagen es un objeto PIL, por lo que no es necesario abrirla nuevamente
    image = image.convert('RGB').resize((224, 224))
    image = np.array(image) / 255.0  # Normalizar la imagen
    image = np.expand_dims(image, axis=0)  # Añadir dimensión de batch

    # Realizar la predicción
    interpreter.set_tensor(input_details[0]['index'], image.astype(np.float32))
    interpreter.invoke()
    output_data = interpreter.get_tensor(output_details[0]['index'])

    # Mostrar el resultado
    return {f"Clase {i+1}": output_data[0][i] for i in range(3)}

# Crear la interfaz de Gradio
interface = gr.Interface(fn=predict,
                         inputs=gr.Image(type="pil"),  # Usar PIL para manejar la imagen
                         outputs=gr.Label(num_top_classes=3))  # Mostrar las 3 clases más probables

# Lanzar la interfaz
interface.launch(debug=True)
```
